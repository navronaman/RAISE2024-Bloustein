# Sentiment Analysis on the Headlines using Roberta Pretrained Model


```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
example = df['title'][70]
print(example)

encoded_text = tokenizer(example, return_tensors='pt')
print(encoded_text)

output = model(**encoded_text)
scores = softmax(output[0][0].detach().numpy())
scores_dict = {
    'neg': scores[0],
    'neu': scores[0],
    'pos': scores[0]
}
print(scores_dict)
```

```python
def polarity_scores(example):
  encoded_text = tokenizer(example, return_tensors='pt')
  output = model(**encoded_text)
  scores = softmax(output[0][0].detach().numpy())
  scores_dict = {
      'title_neg': scores[0],
      'title_neu': scores[0],
      'title_pos': scores[0]
  }
  return scores_dict
```

```python
print(df['title'][1235])
print(polarity_scores(df['title'][1235]))
```

We notice that with headlines, even the expert roBERTa model is unable to make detect the positivity and negativity.

```python
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
  try:
    title = row['title']
    myid = row['ID']
    result = polarity_scores(title)
    res[myid] = result
  except Exception as e:
    print(f'Broke for {myid}')
```

# Let's extract the article texts

```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_url(url):
    try:
        # Fetch the HTML content of the webpage
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for 4xx or 5xx errors
        html_content = response.text

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')

        # Find and extract the text content from the webpage
        text_content = ''
        for paragraph in soup.find_all('p'):
            text_content += paragraph.get_text() + '\n'

        return text_content.strip()  # Strip leading and trailing whitespace

    except Exception as e:
        return f"An error occurred: {e}"
```

```python
example = df['final_redirected_URL'][70]
print(example)
extract_text_from_url(example)
```

## We need a function to summarize the text

```python
import nltk
from gensim.summarization import summarize

# Download NLTK tokenizer models
nltk.download('punkt')

def summarize_article(article_text, word_count=300):
    """
    Summarizes the given article text to a specified word count.

    Args:
    - article_text (str): The full text of the article to be summarized.
    - word_count (int): The desired word count of the summary. Default is 300.

    Returns:
    - summary (str): The summarized text of the article.
    """
    summary = summarize(article_text, word_count=word_count)
    return summary
```
